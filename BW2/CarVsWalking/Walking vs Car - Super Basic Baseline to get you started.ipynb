```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
path = "dataset_5secondWindow.csv"
```

```python
df = pd.read_csv(path, index_col=0)
```

```python
columns = df.columns
columns
```

```python
df = df.dropna(axis=1, how="all").fillna(-1)
grouped_by_target = df.groupby("target")
labels =  pd.concat([group for (name, group) in grouped_by_target if name in ['Car', 'Walking']]).target
features = pd.concat([group for (name, group) in grouped_by_target if name in ['Car', 'Walking']])
features = features.drop(["target","id","user"], axis=1)
```

```python
features
```

```python
labels
```

```python
X, y = features, labels
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
decision_tree = tree.DecisionTreeClassifier(random_state=0, max_depth=50)
```

```python
clf = decision_tree.fit(X_train, y_train)
```

```python
print("Accuracy:",((y_test == "Walking") == (clf.predict(X_test) == "Walking")).sum()/len(y_test))
```
